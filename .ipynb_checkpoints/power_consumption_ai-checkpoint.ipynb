## Checking for Null Values
It looks like there are about 25,979 null values out of 2,075,259. That's approximately 1% of the data. Also, from below I can see that the null data values are all in the same row.

```python
df.isnull().sum()
```

```python
df[df['Global_active_power'].isnull()].head(20)
```

Now, I'll need to decide what to do with the null data. One possibility is to drop the data, but then I would have gaps in the timeseries. Another possibility is to fill in the data with the average. But I don't think that's the best option since maybe the data just before the null values were at a low or at a high. Then filling it with the average would cause an upward/downward spike. I think the best option would be to backfill or forward fill the data. I think this makes the most sense since the data is in chronological order. I am going to choose ffill, which will propogate the last valid observation forward.

```python
df = df.fillna(method='ffill', axis=0)
```

```python
df.isnull().sum()
```

Next, I want to see what some of the data looks like. I created a function to plot the last "timestart" number of points over time. Below, I graphed Voltage and Global_active_power. I'm going to make predictions with Voltage.

## Split into Training and Test Set
I'm going to split the data so that the test set is the last 1000 rows.

```python
# since time samples are taken every minute, 60 minutes * 24 hours = 1440 samples in a day
split_time = 1440

time_train = df.index[:-2 *split_time]
x_train = df['Voltage'][:-2 * split_time]

time_valid = df.index[-2 *split_time: -split_time]
x_valid = df['Voltage'][-2 * split_time:-split_time]

time_test = df.index[-split_time:]
x_test = df['Voltage'][-split_time:]
```

```python
def plot_series(time, series, format="-", start=0, end=None):
    plt.plot(time[start:end], series[start:end], format)
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid(True)
```

```python
plt.figure(figsize=(14,7))
plot_series(time_valid, x_valid)
```

## Naive Forecast
A naive forecast is where the last increment's actuals are used as this increment's forecast. There are no adjustments or attempts to establish causal factors. This is important because it establishes a baseline which I can use to compare with forecasts generated by more sophisticated techniques.

```python
naive_forecast = df['Voltage'][-2 * split_time - 1:-split_time - 1]
```

```python
plt.figure(figsize=(14, 7))
plot_series(time_valid, x_valid, start= 0, end=150)
plot_series(time_valid, naive_forecast, start = 0, end=150)
plt.legend(['Actual','Predicted'])
plt.show()
```

## Computing Metrics on Naive Prediction to get Baseline
Now I'm going to compute the mean squared error and mean absolute error between the forecasts and the predictions in the validation period. This will be my baseline. Ideally, the more sophisticated algorithms should perform better than this.

```python
from tensorflow import keras
print(keras.metrics.mean_squared_error(x_valid, naive_forecast).numpy())
print(keras.metrics.mean_absolute_error(x_valid, naive_forecast).numpy())
```

## Moving Average Forecast
Next, I'll compute the moving average forecast which calculates the mean of the last few values and uses that to forecast the next value. If the window_size=1, then this is equivalent to naive forecast. 

```python
def moving_average_forecast(series, window_size):
    forecast = []
    for time in range(len(series) - window_size):
        forecast.append(series[time:time+window_size].mean())
    return np.array(forecast)
```

```python
win_size = 10
moving_avg = moving_average_forecast(x_valid, win_size)
```

```python
plt.figure(figsize=(14, 7))
plot_series(time_valid, x_valid)
plot_series(time_valid[win_size:], moving_avg)
```

## Checking MSE and MAE of Moving Average
As I can see below, the MSE and MAE is actually worse than the naive forecast

```python
print(keras.metrics.mean_squared_error(x_valid[win_size:],moving_avg).numpy())
print(keras.metrics.mean_absolute_error(x_valid[win_size:],moving_avg).numpy())
```

## Converting into a  WindowDataset
First, I am going to manually do the conversion with a test_dataset and print the some of the output after each step. This way I can see how the dataset changes with each step. Then, I am going to write a function in order to automatically do all of these steps.

```python
# converting to tf.data.Dataset
test_dataset = tf.data.Dataset.from_tensor_slices(x_valid)
dataset_list = list(test_dataset.as_numpy_iterator())
dataset_list[0:10]
```

```python
# converting to WindowDataset, using window_size + 1 so that I can use the last number in the window as the target
window_size = 5
test_dataset = test_dataset.window(window_size + 1, shift = 1, drop_remainder=True)

# I'm going to comment the below code out. This just prints every single window. It was just too long when committing to github
#for window in test_dataset:
    #print(list(window.as_numpy_iterator()))
```

```python
# flattening the result
test_dataset = test_dataset.flat_map(lambda window: window.batch(window_size+1))
list(test_dataset.as_numpy_iterator())[0:10]
```

```python
# shuffling the dataset and mapping the dataset into a tuple containing the input and the actual prediction
shuffle_buffer = 5
test_dataset = test_dataset.shuffle(shuffle_buffer).map(lambda window: (window[:-1], window[-1]))
list(test_dataset.as_numpy_iterator())[0:10]
```

```python
# batching the dataset
batch_size = 5
test_dataset = test_dataset.batch(batch_size).prefetch(5)
list(test_dataset.as_numpy_iterator())[:2]
```

```python
def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
    # converting to tf.data.Dataset
    dataset = tf.data.Dataset.from_tensor_slices(x_train)
    # converting to WindowDataset, using window_size + 1 so that I can use the last number in the window as the target
    dataset = dataset.window(window_size + 1, shift = 1, drop_remainder=True)
    # flattening the result
    dataset = dataset.flat_map(lambda window: window.batch(window_size+1))
    # shuffling the dataset and mapping the dataset into a tuple containing the input and the actual prediction
    dataset = dataset.shuffle(shuffle_buffer).map(lambda window: (window[:-1], window[-1]))
    # batching the dataset
    dataset = dataset.batch(batch_size).prefetch(5)
    return dataset
```

```python
# set window_size to be 60, which would be 1 hour
window_size = 20
batch_size = 32
shuffle_buffer_size = 1000
```

```python
dataset = windowed_dataset(x_train, window_size, batch_size, shuffle_buffer_size)
```

```python
validation_dataset = windowed_dataset(x_valid, window_size, batch_size, shuffle_buffer_size)
```

# NN Forecast

```python
l0 = tf.keras.layers.Dense(1, input_shape=[window_size])
model = tf.keras.models.Sequential([l0])


model.compile(loss="mse", optimizer=tf.keras.optimizers.SGD(lr=1e-6, momentum=0.9))
model.fit(dataset,epochs=2, validation_data = validation_dataset)
```

# Deep NN Forecast

```python
model2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(10, activation="relu", input_shape=[window_size]),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(1)
])
```

### Create callbacks


```python
# learning rate scheduler callback
lr_schedule = tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: 1e-8 * 10**(epoch / 5))

# set up optimizer
optimizer = tf.keras.optimizers.SGD(lr=1e-8, momentum=0.9)

# tensorboard callback
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = log_dir, histogram_freq = 1)

# early stop callback
early_stop = tf.keras.callbacks.EarlyStopping(monitor='loss', 
                                              min_delta=3, 
                                              patience=3, 
                                              verbose=0,
                                              mode='auto', 
                                              baseline=None, 
                                              restore_best_weights=True
                                             )
```

```python
model2.compile(loss="mse", optimizer=optimizer)
history2 = model.fit(dataset,epochs=3, callbacks=[tensorboard_callback], validation_data=(validation_dataset))
```

# RNN Forecast

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Lambda(lambda x: tf.expand_dims(x, axis=-1),
                      input_shape=[None]),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32)),
    tf.keras.layers.Dense(1),
    tf.keras.layers.Lambda(lambda x: x * 100.0)
])
```

```python
model.compile(loss="mse", optimizer=optimizer)
history = model.fit(dataset,epochs=4, callbacks=[tensorboard_callback], validation_data=(validation_dataset))
```

# Convolutional RNN Forecast

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Lambda(lambda x: tf.expand_dims(x, axis=-1), input_shape=[None]),
    tf.keras.layers.Conv1D(filters=32, kernel_size=5, strides=1, padding="causal", activation="relu"),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32, return_sequences=True)),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32, return_sequences=True)),
    tf.keras.layers.Dense(1)
])
```

### Compile and Fit model

```python
model.compile(loss="mse", optimizer=optimizer)
history = model.fit(dataset,epochs=50, callbacks=[lr_schedule, tensorboard_callback, early_stop], validation_data=(validation_dataset))
```

### Plot Learning Rate over time

```python
lrs = 1e-8 * (10 ** (np.arange(17) /20))
plt.semilogx(lrs, history.history['loss'])
plt.axis([1e-8, 1e-7, 0 , 300])
```
